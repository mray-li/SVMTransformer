```python
import re
import matplotlib.pyplot as plt

```

```python
results=[]
for att in ['_svm/']:
    for ds in ['train', 'test']:
        with open('result'+att+ds+'_loss.txt','r') as f:
            file = f.read() 
            file = re.sub('\\[', '', file)
            file = re.sub('\\]', '', file)
            results.append([float(i) for i in file.split(',')])
results
```

```python
for name in ['train','test']:
    with open('saved/transformer-base/'+name+'.txt','r') as f:
        file = f.read() 
        file = re.sub('\\[', '', file)
        file = re.sub('\\]', '', file)
        dt = [float(i) for i in file.split(',')]
        results.append(dt[:len(results[0])])
results
```

```python
plt.figure(figsize=(8, 6), dpi=150)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('training result')
plt.grid(True, which='both', axis='both')
clr = ['red','blue']

for i, name in enumerate(['train','val']):
    plt.plot(results[i], label='svm_'+name, color=clr[i])
for i, name in enumerate(['train','val']):
    plt.plot(results[i+2], label=name, ls='--',color=clr[i])
plt.legend()
plt.show()
```
